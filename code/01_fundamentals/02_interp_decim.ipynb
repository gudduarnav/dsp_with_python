# Interpolation and Decimation

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
import scipy.fftpack
```

A signal x(n) is generated for 1 seconds. Then it is sampled at T1=1/2000 secs.

```python
num_samples = 2000
T1 = 1/2000
t = np.arange(0, num_samples, step=T1)
```

```python
freq = 10
x = np.cos(2*pi*freq*t)
```

```python
plt.figure()
plt.plot(t,x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xlim(t[0], t[num_samples-1])
plt.grid()
```

Plot the DTFT of the above using FFT

```python
fft_x = scipy.fftpack.fft(x)
fft_t = np.linspace(0.0, 1.0/(2.0*1/T1), num_samples//2)
```

```python
plt.figure()
plt.plot(np.abs(fft_x))
```
